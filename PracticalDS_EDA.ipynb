# ***ANALYZING EMS RESPONSE TIME***

**--> Importing Pandas Library**

```python
import pandas as pd
```

**--> Extracting Data File**

```python
data = pd.read_csv("/content/data200k.csv")
```

**--> Dropping Null values**

```python
data = data.dropna()
```

**--> Printing Basic information and Descriptive statistics**

```python
data.info()
data.describe(include='all')
```

**--> Printing the first 5 entries**

```python
data.head(5)
```

**--> Printing the last 5 entries**

```python
data.tail(5)
```

**--> Examining categorical variables for their unique values and frequency distributions**

```python
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(data[col].value_counts())
```

**--> Plotting distributions of numerical columns**

```python
import matplotlib.pyplot as plt
import seaborn as sns
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns.drop('CAD_INCIDENT_ID')
data[numerical_cols].hist(figsize=(15, 10), bins=20)
plt.tight_layout()
plt.show()
```

**--> Correlation heatmap**

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

**--> Converting datetime columns from strings to datetime objects (suitable format for further analysis)**

```python
datetime_cols = ['INCIDENT_DATETIME', 'FIRST_ASSIGNMENT_DATETIME', 'FIRST_ON_SCENE_DATETIME',
                 'FIRST_TO_HOSP_DATETIME', 'FIRST_HOSP_ARRIVAL_DATETIME', 'INCIDENT_CLOSE_DATETIME']
date_format = '%m/%d/%Y %I:%M:%S %p'
for col in datetime_cols:
    data[col] = pd.to_datetime(data[col], format=date_format, errors='coerce')
```

**--> Checking the datetime columns**

```python
data[datetime_cols].head(1)
```

**--> Calculating the time differences between various stages of incident response**

```python
data['Dispatch_to_Scene'] = (data['FIRST_ON_SCENE_DATETIME'] - data['FIRST_ASSIGNMENT_DATETIME']).dt.total_seconds()
data['Scene_to_Hospital'] = (data['FIRST_HOSP_ARRIVAL_DATETIME'] - data['FIRST_ON_SCENE_DATETIME']).dt.total_seconds()
```

**--> Analysis of Response Times**

```python
data[['Dispatch_to_Scene', 'Scene_to_Hospital']].describe()
```

**--> Plotting histogram for frequency distribution of response times**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data['Dispatch_to_Scene'].dropna(), bins=30, kde=True, color='blue')
plt.title("Distribution of Dispatch to Scene Time (seconds)")

plt.subplot(1, 2, 2)
sns.histplot(data['Scene_to_Hospital'].dropna(), bins=30, kde=True, color='green')
plt.title("Distribution of Scene to Hospital Time (seconds)")

plt.tight_layout()
plt.show()
```

**--> Geographical Analysis : Mapping distribution of incidents by borough or ZIP code, using bar plots**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='BOROUGH', order=data['BOROUGH'].value_counts().index)
plt.title("Incidents by Borough")
plt.xticks(rotation=45)
plt.show()
```

**--> Plotting the top 10 different incident types**

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=data['FINAL_CALL_TYPE'], order=data['FINAL_CALL_TYPE'].value_counts().index[:10], palette="coolwarm")
plt.title("Top 10 Finalized Incident Types")
plt.xlabel("Incident Count")
plt.ylabel("Final Call Type")
plt.show()
```

**--> Plotting the severity distribution for the entire dataset**

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='FINAL_SEVERITY_LEVEL_CODE', palette="Spectral")
plt.title("Distribution of Incident Severity Levels")
plt.xlabel("Severity Level")
plt.ylabel("Count")
plt.show()
```

**--> Temporal Patterns of Incidents**

```python
# Creating additional time-related columns
data['INCIDENT_HOUR'] = data['INCIDENT_DATETIME'].dt.hour
data['INCIDENT_DAY'] = data['INCIDENT_DATETIME'].dt.day_name()
data['INCIDENT_MONTH'] = data['INCIDENT_DATETIME'].dt.month

# Plotting incidents by hour of the day
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='INCIDENT_HOUR', palette="plasma")
plt.title("Incidents by Hour of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Incident Count")
plt.show()

# Plotting incidents by day of the week
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='INCIDENT_DAY', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette="viridis")
plt.title("Incidents by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Incident Count")
plt.show()

# Plotting incidents by month
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='INCIDENT_MONTH', palette="cividis")
plt.title("Incidents by Month")
plt.xlabel("Month")
plt.ylabel("Incident Count")
plt.show()
```

**--> Examine if higher-severity incidents tend to have quicker response times**

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=data, x='FINAL_SEVERITY_LEVEL_CODE', y='Dispatch_to_Scene', palette="YlGnBu")
plt.title("Dispatch to Scene Time by Severity Level")
plt.xlabel("Severity Level")
plt.ylabel("Dispatch to Scene Time (seconds)")
plt.show()
```

**--> Kernel Density Estimation (KDE) Plot**

```python
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns.drop('CAD_INCIDENT_ID')
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(data[col], shade=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

**--> Boxplots for Outlier Detection**

```python
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns.drop('CAD_INCIDENT_ID')
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

**--> Scatter Plot for Two Continuous Variables**

```python
# Scatter plot between 'INCIDENT_TRAVEL_TM_SECONDS_QY' and 'INCIDENT_RESPONSE_SECONDS_QY'
plt.figure(figsize=(16, 12))
plt.scatter(data['INCIDENT_TRAVEL_TM_SECONDS_QY'], data['INCIDENT_RESPONSE_SECONDS_QY'], alpha=0.5, color='red')
plt.title('Scatter Plot: INCIDENT_TRAVEL_TM_SECONDS_QY vs INCIDENT_RESPONSE_SECONDS_QY')
plt.xlabel('Incident Travel Time (Seconds)')
plt.ylabel('Incident Response Time (Seconds)')
plt.show()
```

**--> Top 5 precincts with the highest average INCIDENT_TRAVEL_TM_SECONDS_QY**

```python
top_5_precincts_by_travel_time = data.groupby('POLICEPRECINCT')['INCIDENT_TRAVEL_TM_SECONDS_QY'].mean().nlargest(5)
print("Top 5 precincts with highest average INCIDENT_TRAVEL_TM_SECONDS_QY:\n", top_5_precincts_by_travel_time)
```

**--> Top 5 ZIPCODEs with the highest average INCIDENT_RESPONSE_SECONDS_QY**

```python
top_5_zipcodes_by_response_time = data.groupby('ZIPCODE')['INCIDENT_RESPONSE_SECONDS_QY'].mean().nlargest(5)
print("Top 5 ZIPCODEs with highest average INCIDENT_RESPONSE_SECONDS_QY:\n", top_5_zipcodes_by_response_time)
```

**--> Top 10 Highest Correlated Columns**

```python
# Filtering out non-numeric columns
numeric_data = data.select_dtypes(include=[float, int])

# Calculating the correlation matrix only for numeric columns
correlation_matrix = numeric_data.corr()

# Unstacking the correlation matrix to find pairs of columns with the highest correlation
correlations = correlation_matrix.unstack().sort_values(ascending=False)

# Removing self correlations (correlation of a column with itself)
correlations = correlations[correlations < 1]

# Top 10 pairs of columns with the highest correlations
top_10_correlated_pairs = correlations.drop_duplicates().head(10)
print("Top 10 correlated column pairs:\n", top_10_correlated_pairs)
```
